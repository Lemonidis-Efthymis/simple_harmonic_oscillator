```python
import numpy as np
import matplotlib.pyplot as plt

# ---  ---
m = 0.5       #   kg
k = 20.0      #    N/m
x0 = 0.2      #   ()  
t_max = 5.0   #   ()

# ---  ---
#     = sqrt(k/m)
omega = np.sqrt(k / m)

#    0  t_max
t = np.linspace(0, t_max, 500)

#  : x(t) = x0 * cos(*t)
x = x0 * np.cos(omega * t)

# ---   ---
plt.figure(figsize=(10, 5))
plt.plot(t, x, color='blue', linewidth=2, label=' ')

#  
plt.title('   ( -)')
plt.xlabel(' (s)')
plt.ylabel(' x (m)')
plt.axhline(0, color='black', linestyle='--') #  
plt.grid(True, alpha=0.3)
plt.legend()



plt.show()
```
